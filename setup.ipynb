```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

print("numpy:", np.__version__)
print("pandas:", pd.__version__)

X = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])

model = LinearRegression().fit(X, y)

print("coef:", model.coef_, "intercept:", model.intercept_)
```
